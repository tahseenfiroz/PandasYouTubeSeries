```python
import pandas as pd
cric_players = [['Virat Kohli','India','RCB']
            ,['Rohit Sharma','India','MI']
            ,['MS Dhoni','India','CSK']
            ,['Pat Cummins','Australia','SRH']           
           ] 
df_cric_players = pd.DataFrame(cric_players,columns=["PlayerName","Country","IPLTeams"])
df_cric_players
```

```python
import pandas as pd
cric_players_India_role = [['Virat Kohli','Batsmen']
            ,['Rohit Sharma','Bowler']
            ,['MS Dhoni','WicketKeeper']
            ,['Zaheer Khan ','WicketKeeper']
                     
           ] 
df_cric_players_role = pd.DataFrame(cric_players_India_role,columns=["PlayerName","Role"])
df_cric_players_role
```

```python
df_out = df_cric_players.merge(df_cric_players_role,how='inner',left_on='PlayerName',right_on='PlayerName')
df_out
```

```python
df_out = df_cric_players.merge(df_cric_players_role,how='left',left_on='PlayerName',right_on='PlayerName')
df_out
```

```python
df_out = df_cric_players.merge(df_cric_players_role,how='right',left_on='PlayerName',right_on='PlayerName')
df_out
```
